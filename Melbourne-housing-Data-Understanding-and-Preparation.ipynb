```python
# import pandas and numpy librarys

import pandas as pd
import numpy as np
```

```python
# read dataset csv file 

data = pd.read_csv("Melbourne_housing_FULL.csv")
data

#34857 rows × 21 columns
```

### Check Missing

```python
missing = pd.DataFrame({'missing': data.isnull().sum()})
missing
```

### handle Missing using drop and fill (median)

```python
#drop valus with null in Price
price_dropna = data.dropna(axis = 0, subset=['Price'])
price_dropna

# 27247 rows × 21 columns
```

```python
# check NaN Values
data.isnull().sum()
```

```python
for column in ['Rooms','Price', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt','Lattitude', 'Longtitude', 'Propertycount' ]:
       data[column].fillna(data[column].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
for column in [ 'CouncilArea', 'Regionname']:
    data[column].fillna(data[column].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt

atts = list = ['Suburb','Address','Rooms','Type','Price','Method','SellerG','Date','Distance','Postcode','Bathroom'
       ,'Car','Landsize','BuildingArea','YearBuilt','CouncilArea','Regionname','Propertycount']
plt.figure(figsize =(20,10))
data[atts].boxplot()
plt.title('melbourne attributes')
```

```python
# Remove the noise value
```

```python
data.loc[data['Rooms'] < (data['Bedroom2'] + data['Bathroom']), 'Rooms'] = data['Bedroom2'] + data['Bathroom']

data = data[(data['BuildingArea'] >= 70) & (data['BuildingArea'] <= 2500)]
data = data[(data['YearBuilt'] >= 1900) & (data['YearBuilt'] <= 2021)]
data = data[(data['Landsize'] >= 100) & (data['Landsize'] <= 5000)]
data = data[(data['Price'] >= 1000) & (data['Price'] <= 5000000)]

data
```

```python
data[atts].boxplot()
```

```python
# Check duplicate records
# Remove duplicate records
```

```python
dupli_rec = data.duplicated()
data.drop_duplicates(subset=None,inplace=True)
dupli_rec.sum()
```

## Apply discretization on numeric attributes 

```python
bins_price = [57000 , 120000 , 170000 , 600000 , 9000000]
bins_name = ['low','middle','hight','very hight']
data['Price'] = pd.cut(data['Price'], bins_price , labels = bins_name)
```
